```python
def process_world_bank_data(filename):
    # Read the CSV file
    df = filename

    # Convert the wide format to a long format
    df_long = df.melt(id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], var_name='Year', value_name='Value')

    # Pivot the long format to get years as columns
    df_years = df_long.pivot_table(index=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], columns='Year', values='Value').reset_index()

    # Pivot the long format to get countries as columns
    df_countries = df_long.pivot_table(index=['Indicator Name', 'Indicator Code', 'Year'], columns=['Country Name', 'Country Code'], values='Value').reset_index()

    # Clean the transposed dataframes
    df_years.columns.name = None
    df_countries.columns.name = None

    return df_years, df_countries

```

```python
import pandas as pd
import matplotlib.pyplot as plt

def process_world_bank_data(filename, indicator_codes):
    df = filename

    # Filter data for the required indicators
    df_filtered = df[df['Indicator Code'].isin(indicator_codes)]

    # Convert to long format
    df_long = df_filtered.melt(id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], var_name='Year', value_name='Value')

    # Pivot the data to have one row per country and year
    df_pivoted = df_long.pivot_table(index=['Country Name', 'Country Code', 'Year'], columns=['Indicator Name', 'Indicator Code'], values='Value').reset_index()
    df_pivoted.columns = [' '.join(col).strip() for col in df_pivoted.columns.values]

    return df_pivoted

def plot_co2_vs_gdp(df):
    # Drop rows with missing values
    df_clean = df.dropna()

    # Create the scatter plot
    plt.scatter(df_clean['GDP per capita (constant 2010 US$) NY.GDP.PCAP.KD'], df_clean['CO2 emissions (metric tons per capita) EN.ATM.CO2E.PC'], alpha=0.5)
    plt.xlabel('GDP per capita (constant 2010 US$)')
    plt.ylabel('CO2 emissions (metric tons per capita)')
    plt.title('CO2 Emissions vs. GDP per capita')
    plt.grid(True)
    plt.show()

filename = df4
indicator_codes = ['NY.GDP.PCAP.KD', 'EN.ATM.CO2E.PC']  # GDP per capita (constant 2010 US$) and CO2 emissions (metric tons per capita)

df = process_world_bank_data(filename, indicator_codes)
plot_co2_vs_gdp(df)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def process_world_bank_data(filename, indicator_codes):
    df = df4

    # Filter data for the required indicators
    df_filtered = df[df['Indicator Code'].isin(indicator_codes)]

    # Convert to long format
    df_long = df_filtered.melt(id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], var_name='Year', value_name='Value')

    # Pivot the data to have one row per country and year
    df_pivoted = df_long.pivot_table(index=['Country Name', 'Country Code', 'Year'], columns=['Indicator Name', 'Indicator Code'], values='Value').reset_index()
    df_pivoted.columns = [' '.join(col).strip() for col in df_pivoted.columns.values]

    return df_pivoted

def plot_methane_and_other_parameters(df):
    # Drop rows with missing values
    df_clean = df.dropna()

    # Create a pair plot of the indicators
    indicators = ['Access to electricity (% of population) EG.ELC.ACCS.ZS',
                  'Methane emissions (% change from 1990) EN.ATM.METH.ZG',
                  'Methane emissions (kt of CO2 equivalent) EN.ATM.METH.KT.CE',
                  'Nitrous oxide emissions (% change from 1990) EN.ATM.NOXE.ZG',
                  'Nitrous oxide emissions (thousand metric tons of CO2 equivalent) EN.ATM.NOXE.KT.CE']
    sns.pairplot(df_clean[indicators])
    plt.show()

filename = df4
indicator_codes = ['EN.ATM.METH.ZG', 'EN.ATM.NOXE.ZG', 'EN.ATM.NOXE.KT.CE', 'EN.ATM.METH.KT.CE', 'EG.ELC.ACCS.ZS']

df = process_world_bank_data(filename, indicator_codes)
plot_methane_and_other_parameters(df)
```
